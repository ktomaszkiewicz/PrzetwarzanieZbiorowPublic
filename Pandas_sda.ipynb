```python
#chcemy ustawic inny indeks\
#albo to robimy na poziomie wczytywania danych
movie = pd.read_csv('data/movie.csv',index_col = 'movie_title')

```

```python
movie.columns
```

```python
#zmiana indexu
#reset indexu
movie = movie.reset_index()
#stary indeks stal sie kolumna
```

```python
movie.set_index('director_name')
#uwaga teraz index sie powtarza np. steven spielberg
```

```python
#sortowanie df  wtedy indeks nie jest w poprzedniej kolejnosci
movie = movie.sort_values('duration')
movie.reset_index() #stary indeks staje sie kolumna
movie.reset_index(drop=True) #nie ustawia starego indeksu jako nowa kolumne
```

```python
#sortowanie po dwoch kolumanch
movie.sort_values(['director_name','duration'],ascending = False)
```

```python
#powracanie do poprzedniej wersji czyli posortowac po indeksie\
movie.sort_index(ascending = False)
#sortowanie po kolumnach bo wybralam nowe axis =1
movie.sort_index(ascending = False, axis = 1)
```
