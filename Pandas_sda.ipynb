# merge, join , concate

```python
df1 = pd.DataFrame(
        {
            "A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
            "C": ["C0", "C1", "C2", "C3"],
            "D": ["D0", "D1", "D2", "D3"],
        },
        index=[0, 1, 2, 3],
    )
    

df2 = pd.DataFrame(
        {
            "A": ["A4", "A5", "A6", "A7"],
            "B": ["B4", "B5", "B6", "B7"],
            "C": ["C4", "C5", "C6", "C7"],
            "D": ["D4", "D5", "D6", "D7"],
        },
        index=[4, 5, 6, 7],
    )
    

df3 = pd.DataFrame(
        {
            "A": ["A8", "A9", "A10", "A11"],
            "B": ["B8", "B9", "B10", "B11"],
            "C": ["C8", "C9", "C10", "C11"],
            "D": ["D8", "D9", "D10", "D11"],
        },
        index=[8, 9, 10, 11],
    )
```

```python
from IPython.display import display
display(df1)
display(df2)
display(df3)
```

```python
results = pd.concat([df1,df2,df3])
results
```

```python
#dodatkowo rozroznic z ktorego setu co pochodzi
#multiindex w pandasie - doczytac o tym
results = pd.concat([df1,df2,df3], keys = ['x','y','z'])
```

```python
results.loc['y']
```

```python
df4 = pd.DataFrame(
        {
            "B": ["B2", "B3", "B6", "B7"],
            "D": ["D2", "D3", "D6", "D7"],
            "F": ["F2", "F3", "F6", "F7"],
        },
        index=[2, 3, 6, 7],
    )
```

```python
pd.concat([df1,df4])
```

# funkcja apply

```python
#skomplikowane operacje - stworzyc nowa kolumne ktora mowi jakie jest film(dlugi, krotki, sredni)
movie['duration'].apply(np.sqrt)
```

```python
movie_1 = movie[['duration','num_critic_for_reviews']]
movie_1
```

```python
#przypisanie pewnych wartosci w zaleznosci od innej kolumny
def movie_type(x):
    if x > 120:
        return 'dlugi'
    elif x < 60:
        return 'krotki'
    else:
        return 'sredni'
    
movie_1['duration'].apply(movie_type)
movie_1
```

```python
movie_1.apply(lambda x: movie_type(x['duration']), axis = 1)
```

```python
movie_1.apply(lambda x: print(x), axis = 1)
```

# group by

```python
data_path = 'https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/flights.csv'
flights = pd.read_csv(data_path)
flights.head()
```

```python
flights.groupby('AIRLINE').agg({'ARR_DELAY': 'mean'})
```

```python
flights.groupby('AIRLINE').agg('mean')
```

```python
flights.groupby('AIRLINE')[['ARR_DELAY']].mean()
```

```python
flights.groupby('AIRLINE').agg({'ARR_DELAY': 'mean', 'DEP_DELAY': 'sum'})
```

```python
##srednie opoznienie w kazdym dniu
flights.groupby(['AIRLINE','WEEKDAY']).agg('mean')
```

```python
fdf = flights.groupby(['AIRLINE','WEEKDAY'])[['DEP_DELAY','ARR_DELAY']].agg('mean')
```

```python
fdf.loc['AA']
```

```python
flights.groupby(['AIRLINE','WEEKDAY'])[['DEP_DELAY','ARR_DELAY']].agg(['mean','count'])
```
